1. ¿Cuántas películas tienen una duración superior a 120 minutos?

```python
mycursor.execute('''SELECT COUNT(idFilm) as amount 
FROM films
WHERE runtime >= 120''')
myresult = mycursor.fetchone()
myresult
```

2. ¿Cuántas películas incluyen subtítulos en español?

```python

mycursor.execute("""SELECT COUNT(films.idFilm) as amount
                    FROM films
                    INNER JOIN subtitles_films ON films.idFilm = subtitles_films.idFilm
                    INNER JOIN languages ON languages.idLang = subtitles_films.idLang
                    WHERE languages.language = 'es'"""
                 )
myresult = mycursor.fetchone()
myresult
```

3. ¿Cuántas películas tienen contenido adulto?

```python
mycursor.execute("""SELECT COUNT(idFilm) as amount
                    FROM films
                    WHERE rating IS true"""
                 )
myresult = mycursor.fetchone()
myresult
```

4. ¿Cuál es la película más antigua registrada en la base de datos?

```python
mycursor.execute("""SELECT title , year
FROM films
ORDER BY year asc LIMIT 1""")
myresult = mycursor.fetchall()
myresult
```

5. Muestra el promedio de duración de las películas agrupado por género.

```python


mycursor.execute("""SELECT genre, AVG(runtime) as AverageRuntime
                    FROM films
                    GROUP BY genre"""
                 )
myresult = mycursor.fetchall()
myresult
```

6. ¿Cuántas películas por año se han registrado en la base de datos? Ordena de mayor a menor.

```python
mycursor.execute("""SELECT year, COUNT(idFilm) as amount
                    FROM films
                    GROUP BY year
                    ORDER BY amount desc"""
                 )
myresult = mycursor.fetchall()
myresult
```

7. ¿Cuál es el año con más películas en la base de datos

```python
mycursor.execute("""SELECT year, COUNT(idFilm) as amount
                    FROM films
                    GROUP BY year
                    ORDER BY amount desc LIMIT 1;"""
                 )
myresult = mycursor.fetchall()
myresult
```
